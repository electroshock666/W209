```python
# 
# Credit to Chris McClure-St. Amant for the notebook used in this example (https://colab.research.google.com/drive/1UzUa_tj2ScZSCCsOhOE2rnq8nXKbzk_h?authuser=2)
#
import pandas as pd
from google.colab import files
import numpy as np
pd.__version__
import altair as alt
from altair import datum
```

```python
# instead of requiring the file to be uploaded or loaded from google drive (which won't work for other accounts) it is recommended to use github. 
# all of the data files for W209 are located in the repo below or you can upload to your own repo. 
# *you need the RAW file URL which can be obtained by navigating to the file in the repo on github.com, clicking the '...' at the top right and selecting "copy permalink"
df = pd.read_csv('https://raw.githubusercontent.com/electroshock666/MIDS-W209/6a84d1a5da4952feccc4e0ff85d1198a5b730837/Week_6/players_20.csv')
```

```python
df.head()
```

```python
# Scale back the dataframe to just have the columns we are using
df = df[['long_name','work_rate', 'skill_long_passing', 'skill_fk_accuracy']]
df.columns = ['Name', 'Work Rate', 'Long Passing Skill', 'Free Kick Accuracy']
```

```python
df.head()
```

```python
# Remove 5000 max rows limit 
alt.data_transformers.disable_max_rows()
```

```python
alt.Chart(df, title="Free kick accuracy vs. long passing skill").mark_circle().encode(
    x = alt.X('Free Kick Accuracy:Q', title="Free kick accuracy"),
    y = alt.Y('Long Passing Skill:Q', title="Long passing skill"),
    color = alt.Color('Work Rate:N', scale = alt.Scale(scheme='category10')),
    tooltip = alt.Tooltip(['Name:N', 'Work Rate:N', 'Long Passing Skill:Q', 'Free Kick Accuracy:Q'])
 
).properties(height=600, width=900).interactive()
```
