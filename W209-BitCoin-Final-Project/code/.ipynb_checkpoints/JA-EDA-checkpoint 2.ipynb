```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('bitcoindataset.csv')
```

```python
df.describe()
```

```python
df.head(10)
```

```python
df.drop('Unnamed: 0', 1)
df.set_index("Date", inplace = True)
data = df.reset_index().melt('Date')
data.head(100)
```

```python
alt.Chart(data).mark_line().encode(
    x='x',
    y='value',
    color='variable'
)

chart = alt.Chart(df, title = "BTC Price").mark_point().encode(
    x='Date',
    y='value',
    color='variable',
)

chart.encoding.x.title = 'Date'
chart.encoding.y.title = 'Stock Price'

chart
```
